# Análisis de Datos - Sensor de Stock IoT

Este notebook cumple con el requisito de análisis de datos para el proyecto de IoT. 
Simula el análisis del historial de movimientos de inventario.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Simulación de carga de datos
data = {
    'date_time': ['2024-05-20 08:00:00', '2024-05-20 09:15:00', '2024-05-20 17:30:00', '2024-05-21 08:05:00', '2024-05-21 18:00:00'],
    'product_id': ['PROD-001', 'PROD-002', 'PROD-001', 'PROD-003', 'PROD-002'],
    'movement_type': ['entrada', 'salida', 'salida', 'entrada', 'entrada']
}
df = pd.DataFrame(data)
df['date_time'] = pd.to_datetime(df['date_time'])
df.head()
```

## 1. Frecuencia de Tipos de Movimientos

```python
counts = df['movement_type'].value_counts()
counts.plot(kind='bar', color=['#00CC96', '#EF553B'])
plt.title('Entradas vs Salidas')
plt.show()
```

## 2. Análisis Temporal (Horas Pico de Movimiento)

```python
df['hora'] = df['date_time'].dt.hour
sns.histplot(df['hora'], bins=24, kde=True)
plt.title('Distribución de Movimientos por Hora del Día')
plt.xlabel('Hora')
plt.show()
```
